# Subplots
It is used to show multiple plots side by side in one figure.

## 1. `plt.subplot()`
The syntax is `plt.subplot(rows, cols, index)`

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=43)
x = np.arange(10)
y1 = rng.integers(10, 50, 10)
y2 = rng.integers(20, 60, 10)
y3 = rng.integers(30, 70, 10)
y4 = rng.integers(40, 80, 10)
```

```python
# Basic example

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 1')

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='orange', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 2')

plt.suptitle('This is main title') # Title for overall plot
plt.tight_layout() # This automatically adjust spacing between two plots prevents plot overlaping
plt.show()
```

```python
# 2 rows 2 cols
# 4 plots in one fig

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='red', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y1, color='orange', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y1, color='green', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y1, color='pink', alpha=0.4, linestyle='--', marker='o')
plt.title('plot 4')

plt.suptitle('overall title')
plt.tight_layout()
plt.show()
```
