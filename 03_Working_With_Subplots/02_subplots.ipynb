## 2. `plt.subplots()`
`plt.subplots()` returns a figure and axes objects. Gives more control

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=63)
x = np.arange(10)
y1 = rng.integers(10, 50, 10)
y2 = rng.integers(20, 60, 10)
y3 = rng.integers(30, 70, 10)
y4 = rng.integers(40, 80, 10)
```

```python
# Basic example

fid, axes = plt.subplots(1, 2)

axes[0].plot(x, y1, label='1', color='red')
axes[0].set_title('first')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x, y2, label='2', color='green')
axes[1].set_title('second')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

plt.tight_layout()
# plt.suptitle('Title', fontsize=20)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y1, color='red')
axes[0, 0].set_title('1')
axes[0, 1].plot(x, y1, color='green')
axes[0, 1].set_title('2')
axes[1, 0].plot(x, y1, color='pink')
axes[1, 0].set_title('3')
axes[1, 1].plot(x, y1, color='orange')
axes[1, 1].set_title('4')

plt.tight_layout()
plt.show()
```

`plt.subplot()` - Old way  
`plt.subplots()` - New way

### Share axis
If we want all plots on same axis we can do

```python
fig, axes = plt.subplots(1, 2, sharex=True, sharey=True)

axes[0].plot(x, y1)
axes[1].plot(x, y2)

plt.tight_layout()
plt.show()
```
