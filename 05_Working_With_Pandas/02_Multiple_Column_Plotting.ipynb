# Multiple Column Plotting
When the DataFrame has multiple columns, pandas give ppowerfull way to plot them all together. grouped, stacked or overlapped.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month'   : ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales'   : [1500, 1800, 1200, 2200, 2500, 2100],
    'Profit'  : [300,  400,  250,  500,  600,  450],
    'Expense' : [1200, 1400, 950, 1700, 1900, 1650]
}

df = pd.DataFrame(data)
df.set_index('Month', inplace=True)
df
```

## Plot all columns - one line

```python
df.plot()
plt.show()
```

Pandas automatically add different colors to different columns and also adds legend.

```python
df.plot(kind='bar')
plt.show()
```

All columns appear side by side of each month automatically.

```python
df.plot(kind='bar', stacked=True)
plt.show()
```

```python
df[["Sales", "Profit"]].plot(kind='bar')
plt.show()
```

```python
df.plot(kind='bar',
       subplots=True)
plt.show()
```

```python
df.plot(kind='bar',
       subplots=True,
       layout=(1, 3))
plt.tight_layout()
plt.show()
```
