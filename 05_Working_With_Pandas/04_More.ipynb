# More

we can use   

`df.plot.line()`  like this which is easy to understand and use,

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month'  : ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales'  : [1500, 1800, 1200, 2200, 2500, 2100],
    'Profit' : [300,  400,  250,  500,  600,  450]
}

df = pd.DataFrame(data)
df.set_index('Month', inplace=True)
```

```python
df.plot.line()
```

```python
df.plot.bar()
```
