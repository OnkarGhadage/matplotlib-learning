# Plot directly from Pandas
Pandas have ploting built in, it uses matplotlib under the hood.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Month'  : ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales'  : [1500, 1800, 1200, 2200, 2500, 2100],
    'Profit' : [300,  400,  250,  500,  600,  450]
}

df = pd.DataFrame(data)
df
```

## Simple plot from DataFrame

```python
df.plot()
plt.title('ploted from pd')
plt.show()
```

The months column contain object (str) the pandas exclude that data cause its non numeric. And plot only numeric data.  
Pandas take index as a x axis.  

By two methods we can make Months as a x axis.  
`df.plot(x='month')` and `df.set_index('Month', inplace=True)`

```python
df.plot(x='Month')
plt.title('ploted from pd')
plt.show()
```

```python
df.set_index('Month', inplace=True)
df.plot()
plt.title('ploted from pd')
plt.show()
```

## Ploting single column

```python
df['Sales'].plot()
plt.title('Single column plot')
plt.show()
```

## Choose plot type (`kind='type'`)

```python
df['Sales'].plot(kind='line')
```

```python
df['Sales'].plot(kind='bar')
```

In this way we can plot different types of plot directly from pandas.  

```python
df.plot(x='Sales', y='Profit')
```

```python
df.plot(kind='area')
```
