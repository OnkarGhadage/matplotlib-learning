# Histogram From Pandas
Pandas make histogram easier to plot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Maths'   : [45, 55, 60, 65, 70, 70, 75, 80, 85, 90, 90, 95],
    'Science' : [50, 55, 55, 60, 65, 70, 75, 75, 80, 85, 90, 92],
    'English' : [40, 50, 60, 60, 65, 70, 70, 75, 80, 85, 88, 95]
}

df = pd.DataFrame(data)
df
```

## Single column

```python
df["Maths"].hist()
plt.show()
```

## Custamization of histogram is same as others.

## All column in one plot

```python
df.hist()
plt.show()
```

It autmatically creates subplots

## Overlapping plots

```python
df["Maths"].hist()
df["Science"].hist()
df["English"].hist()
plt.show()
```

# Important

df.hist() - saperate plots + have less control  
df.plot(kind='hist') - All in one chart + have more control
