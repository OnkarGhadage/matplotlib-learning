## Understand `matplotlib.pyplot`

`matplotlib` -> Toolbox  
`pyplot` -> Tool in that toolbox (Gives simple way to create charts and plots)  

### Import pyplot
`import matplotlib.pyplot` OR `import matplotlib.pyplot as plt`

```python
import matplotlib.pyplot as plt
```

```python
# First plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 40])
plt.show()
```

## Basic plot creation

plot is a normal graph with x (horizontal) and y (vertical) axis.

```python
x = [1, 2, 3, 4, 5] # X axis elements
y = [10, 20 ,15, 30 ,40] # Y axis elements
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Title")
plt.xlabel("Time")
plt.ylabel("Spees")
plt.show()
```
