## Using a style

```python
plt.style.use('fivethirtyeight')       # apply style — one line!

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, marker='o')
plt.title("ggplot Style")
plt.show()
```

1. 'ggplot' -> clean, colorful, great for presentation
2. 'seaborn-v0_8' -> modern, minimal, data science favourite
3. 'fivethirtyeight' -> Bold, high contrast
4. 'dark_background' -> dark theme
5. 'bmh' -> clean academic style
