# Log scale
A log scale is used when data spreaded in a very wide range. Like from 1 to 1,000,000.  
On a normal scale, small values get squished and are impossible to read.  
Log scale fixes that.

## Both axes in log scale

```python
x = [1, 10, 100, 1000, 10000]
y = [2, 40, 800, 16000, 320000]

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='tomato', linewidth=2)

plt.xscale('log')     # x in log
plt.yscale('log')     # y in log

plt.title("Both Axes Log Scale")
plt.xlabel("X (log)")
plt.ylabel("Y (log)")
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.show()
```

## Ex

```python
years      = [1800, 1850, 1900, 1950, 1970, 1990, 2000, 2010, 2023]
population = [1000, 1200, 1600, 2500, 3700, 5300, 6100, 6900, 8000]

plt.figure(figsize=(12, 4))

# Normal scale
plt.subplot(1, 2, 1)
plt.plot(years, population, marker='o', color='steelblue', linewidth=2)
plt.title("Normal Scale")
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.grid(True, linestyle='--', alpha=0.7)

# Log scale
plt.subplot(1, 2, 2)
plt.plot(years, population, marker='o', color='orange', linewidth=2)
plt.yscale('log')
plt.title("Log Scale — Growth Rate Clearer")
plt.xlabel("Year")
plt.ylabel("Population (log)")
plt.grid(True, which='both', linestyle='--', alpha=0.7)

plt.suptitle("World Population Growth", fontsize=14)
plt.tight_layout()
plt.show()
```

## Types of scale

`plt.xscale('linear')`   # default — normal scale  
`plt.xscale('log')`      # log base 10  
`plt.xscale('symlog')`   # log scale that handles negative values too  
`plt.xscale('logit')`    # for probability data between 0 and 1
