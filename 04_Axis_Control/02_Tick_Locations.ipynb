# Tick Locations
Ticks are the small marks on the axis that show values.  
By default matplotlib places them automatically and we can custamize it.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed = 84)
x = np.arange(20)
y1 = rng.integers(100, 200, 20)
y2 = rng.integers(150, 250, 20)
```

## Normal plot

```python
# Normal plot
plt.plot(x, y1)
plt.show()
```

In above the default plot is ploted.  
Now lets see manually placing ticks.

## Ploting tick manually

```python
# ploting manually ticks
plt.plot(x, y1, color="tomato", marker='o', markersize=3, alpha=0.4)
plt.xlim(0, 20)
plt.ylim(100, 200)
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([100, 110, 120, 130, 140, 150])
plt.show()
```

## Ploting tick with help of numpy

```python
# With help of numpy
plt.plot(x, y1, color="tomato", marker='o', markersize=3, alpha=0.4)
plt.xlim(0, 20)
plt.ylim(100, 200)
plt.xticks(np.arange(20))
plt.yticks(np.arange(100, 200, 10))
plt.show()
```

## Replacing tick number with label

```python
# Replacing tick numbers with labels
plt.plot(x, y1, color="tomato", marker='o', markersize=3, alpha=0.4)
plt.xlim(0, 20)
plt.ylim(100, 200)
plt.xticks([0, 1, 2, 3, 4, 5, 6], ["a","b","c","d","e","f", "g"])
plt.yticks([100, 110, 120, 130, 140, 150])
plt.show()
```

## Tick customization

**Tick Customizing**  
|parameter|info|ex|
|-|-|-|
|fontsize|tick font size|12, 22|
|color|tick color|'red', '#FFFFFF'|
|rotation|tick rotation for long text|45, 90,..|

```python
plt.plot(x, y1, color="steelblue", marker='o', markersize=3, alpha=0.4)
plt.xlim(0, 20)
plt.ylim(100, 200)
plt.xticks(np.arange(20), color='pink')
plt.yticks(np.arange(100, 200, 10), color='orange', rotation=90)
plt.show()
```

## Hiding ticks completely

```python
# Hiding tick completely
plt.plot(x, y1, color="tomato", marker='o', markersize=3, alpha=0.4)
plt.xlim(0, 20)
plt.ylim(100, 200)
plt.xticks([])
plt.yticks([])
plt.show()
```

## Ticks controlling in Subplots

```python
# While working with subplots

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].plot(x, y1)
axes[0].set_title('1')
axes[0].set_xticks([1, 2, 3, 4])
axes[0].set_xticklabels(['a', 'b', 'c', 'd'])

axes[1].plot(x, y2)
axes[1].set_title('2')
axes[1].set_yticks(np.arange(150,250,10))


plt.show()
```

**With subplots use `axes[0].set_xticks()` and `axes[0].set_xticklabels()` separately.**
