# Saving Plots

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, marker='o', color='steelblue')
plt.title("My First Saved Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', alpha=0.7)

plt.savefig('my_plot.png')    # saves in current folder
plt.show()
```

# IMP 
Always call `plt.savefig()` **before** `plt,show()`.  
After `plt.show()` the plot is cleared the saved image will be blank image.

## DPI control

plt.savefig('name.png', dpi=72)

|DPI|Use|
|-|-|
|72|Quick preview, web|
|100|Default matplotlib|
|150|Presentation|
|300|Reports, printing|
|600|Publication quality|

## File format

|format|for|
|-|-|
|.png|General use|
|.pdf|Reports, Research papers|
|.svg|Web, dashboards|
|.jpg|Space saving|

## Important parameters

1. `bbox_inches='tight'` -> without this, titles and labels often gets cut off at edge.
2. `transparent=True` -> removes white background
3. `facecolor='red', '#FFFFFF'` -> set background color

## Save in specific folder

`plt.savefig('path/file_name.png')`
