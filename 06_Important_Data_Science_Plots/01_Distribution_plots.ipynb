# Distribution plots
In Data science, before building any model we first need to understand the **Data**.  
Data is spread, normal or skewed and are there any outliers.

1. Shape     — is it bell shaped, skewed left, skewed right?
2. Center    — where is the mean and median?
3. Spread    — how wide is the data?
4. Outliers  — are there extreme values?

## Basic distribution plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = {
    'Marks' : np.random.normal(loc=70, scale=10, size=200)
} # Normal distribution - mean=70, std=10
```

### `np.random.normal(loc=70, scale=10, size=200)` this gererates 200 random values with mean 70 and std 10

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['Marks'].hist(bins=20,
                color='steelblue',
                edgecolor='black',
                alpha=0.4,
                figsize=(9, 5))
plt.title("Distribution of student marks")
plt.xlabel('Marks')
plt.show()
```

```python
mean = df["Marks"].mean()
median = df["Marks"].median()
df['Marks'].hist(bins=20,
                color='steelblue',
                edgecolor='black',
                alpha=0.4,
                figsize=(9, 5))
plt.axvline(mean, color='red')
plt.axvline(median, color='orange')
plt.title("Distribution of student marks")
plt.xlabel('Marks')
plt.show()
```

## Important
### If mean ≈ median -> Data is symmetric.
### If they differ -> Data is ckewed.

```python
normal_data = np.random.normal(loc=70, scale=10, size=300)
right_skewed = np.random.exponential(scale=20, size=300)
left_skewed = 100 - np.random.exponential(scale=15, size=300)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, data, title, color in zip(
    axes,
    [normal_data, right_skewed, left_skewed],
    ['Normal Dist', 'Right Skewed', 'Left Skewed'],
    ['steelblue', 'orange', 'tomato']
):
    ax.hist(data, bins=30, color=color, edgecolor='black', alpha=0.4)
    ax.axvline(data.mean(), color='red', linestyle='--')
    ax.axvline(np.median(data), color='tomato', linestyle='--')
    ax.set_title(title)

plt.suptitle('Types of Dist')
plt.tight_layout()
plt.show()
```

In right skewed data mean > median. In left skewed data mean < median. This tells you a lot about your feature before modeling.
