# Horizontal Bar Chart

We discussed this in Bar chart notebook also.

This similat to bar chart just the bars are horizontal.  
Bars go left to right with respect to value. 

```python
import matplotlib.pyplot as plt
```

**Simple plot**

```python
categories = ['Mango', 'Apple', 'Banana', 'Orange']
values     = [30, 20, 25, 15]

plt.barh(categories, values)
plt.show()
```

As bar chart we can customize this also.

---

|parameter|info|ex|
|-|-|-|
|color|bar color|'red', '#FFFFFF'|
|edgecolor|border color of bar|'red', '#FFFFFF'|
|height|thickness|1,3|
|alpha|transparancy|0.0 to 1.0|

```python
# EX
# Sorting the bars
sorted_pairs = sorted(zip(values, categories))
sorted_values, sorted_categoories = zip(*sorted_pairs)
```

```python
plt.barh(sorted_categoories, sorted_values)
plt.show()
```
