# Scatter plot

The Scatter plot shows individual points without connecting them with a line or anything.  
It is same as line plot just lines are missing...

```python
import matplotlib.pyplot as plt
```

**Basic plot**

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 40]
```

```python
plt.scatter(x, y)
plt.show()
```

Just pure dots no connecting lines.  
We can custamize it as well.

---

|parameter|info|ex|
|-|-|-|
|color|dot color|'red', 'green', '#FFFFFF'|
|s|dot size|50, 100|
|marker|dot shape|'*', 'o', '^'|
|alpha|transparancy|0 to 1.0|

```python
# EX
plt.scatter(x,
           y,
           color='black',
           s=200,
           marker='^',
           alpha=0.5)
plt.title("Scatter plot")
plt.xlabel("abc")
plt.ylabel("xyz")
plt.show()
```

```python
# EX2
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_marks  = [40, 50, 45, 65, 70, 75, 85, 95]

plt.scatter(study_hours, exam_marks,
            color='blue',
            s=200,
            marker='*',
            alpha=0.7)

plt.title("Study Hours vs Exam Marks")
plt.xlabel("Study Hours")
plt.ylabel("Exam Marks")
plt.show()
```

```python
# EX3
height = [150, 160, 165, 170, 175, 180, 185]
weight = [50,  60,  65,  70,  72,  80,  85]
```

```python
plt.scatter(height,
           weight,
           color='red',
           s=100,
           marker='+',
           alpha=0.2)
plt.title("Name")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```
