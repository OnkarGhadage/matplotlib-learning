# Histogram

A histogram shows the distribution of data.  
How often values fall within certain ranges.  
A histogram works with numerical data and groups them into bins.

```python
import matplotlib.pyplot as plt
```

Basic plot

```python
data = [10, 20, 20, 30, 30, 30, 40, 40, 50, 60, 60, 70, 80, 90, 90, 90]

plt.hist(data)
plt.show()
```

mpl automatically decides how many bins to use.  
Each bin is a range of values, and the bar height shows how many values fall in that range.

**Understand bins:**  
bins control how many groups your data is divided into.

```python
# Play with bins.
data = [10, 20, 20, 30, 30, 30, 40, 40, 50, 60, 60, 70, 80, 90, 90, 90]

plt.hist(data, bins=9)
plt.show()
```

Custamizing the histogram:

---

|parameter|info|ex|
|-|-|-|
|bins|no. of groups|5, 10, 30|
|color|bar color|'red', '#FFFFFF'|
|edgecolor|color of border of bar|'red', '#FFFFFF'|
|alpha|transparency|0.0 to 1.0|
|density|shows % instead of count|True/False|

```python
# EX1
marks = [45, 50, 55, 55, 60, 60, 65, 65, 65, 70,
         70, 70, 75, 75, 80, 80, 85, 90, 95, 100]

plt.hist(marks,
         bins=10,
         color='orange',
         edgecolor='black',
         alpha=0.8)

plt.title("Student Exam Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.axvline(x=60, label='passing marks', linewidth=2, color='green') # line on chart
plt.legend() # used to get the name on chart for green line
plt.show()
```

`plt.axvline()` draws a vertical line.  
`plt.axhline()` draws a horizontal line.

**Overlapping two histograms**

```python
# EX2
class_A = [45, 55, 60, 65, 70, 70, 75, 80, 85, 90]
class_B = [50, 60, 60, 65, 65, 70, 75, 75, 80, 95]

plt.hist(class_A, color='orange', edgecolor='black', alpha=0.4, label="Class A", bins=2)
plt.hist(class_B, color='red', edgecolor='black', alpha=0.4, label="Class B", bins=2)
plt.legend()
plt.show()
```
