# Plot Customization

We have seen plot custamization while learning about various plots.  
Now we discuss some remaning points in Plot Customization.  

1. Grid (`plt.grid`)

2. Figure size (`plt.figure(figsize=...)`)

## Grid
A grid adds referance lines in the background.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng(seed=55)
x = np.arange(10)
y = rng.integers(50,150,10)
```

```python
plt.plot(x,y, marker='o', color='red', alpha=0.4)
plt.grid(True)
plt.title('chart with grid lines')
plt.show()
```

Grid custamization

---

|parameter|info|ex|
|-|-|-|
|True/False|On/Off grid|True/False|
|color|color of grid lines|'pink', '#FFFFFF'|
|linestyle|line style|'-', '--' ':'|
|linewidth|Width of grid lines|1, 2|
|alpha|transparancy of line|0.0 to 1.0|
|axis|which axis want grid lines|'x'/'y'/'both'|

```python
plt.plot(x,
        y,
        color = 'red',
        alpha = 0.4,
        linestyle = '-.',
        linewidth = 2,
        marker = 'o',
        markersize=10)
plt.grid(True,
        color='green',
        linestyle=':',
        linewidth=1,
        axis='both')
```

## Figure Size
By default matplotlib makes a small plot. `plt.figure()` gives us that control to control image size.

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,marker='o')
plt.show()
```

The `plt.figure()` is called before drawing anything

figsize = (width, height)  
The size is always in inches

`dpi` - controls resolution of plot image

```python
plt.figure(figsize=(10, 5), dpi=72)
plt.plot(x, y, marker = 'o')
plt.show()
```

|dpi|usecase|
|-|-|
|72|Screen display|
|100|Default|
|150|Presentation|
|300|Print qality|
