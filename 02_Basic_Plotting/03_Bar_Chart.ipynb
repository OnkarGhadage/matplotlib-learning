# Bar Chart

Bar chart displays data as a rectangle bars.  
Used for comparing two values.

```python
import matplotlib.pyplot as plt
```

**Basic plot**

```python
categories = ['Mango', 'Apple', 'Banana', 'Orange', 'Guava']
values     = [30, 20, 25, 15, 5]
```

```python
plt.bar(categories, values)
plt.show()
```

So here each catagory gets a bar to represents its value. And all the bars are vertical in this.

We also have horizontal bar chart

```python
plt.barh(categories, values)
plt.show()
```

Both are same just the axis are swapped cause in some cases the horizontal bars are more redable.

We can custamize this chart also.  

---

|parameter|info|ex|
|-|-|-|
|color|color of bar|'red', 'green', '#FFFFFF'|
|width|width of bar|0 to 1.0 (0.8 is default)|
|edgecolor|color of bar edge|'red', 'green', '#FFFFFF'|
|alpha|transparancy|0 to 1.0|

```python
# EX
plt.bar(categories,
       values,
       color = 'orange',
       width = 0.5,
       edgecolor = 'red',
       alpha = 0.2)
plt.title("Fruits")
plt.xlabel("Fruit")
plt.ylabel("Value")
plt.show()
```

```python
# EX 2
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales  = [1500, 1800, 1200, 2200, 2500, 2100]

colors = ['red', 'black', 'orange', 'yellow', 'green']

plt.bar(months, sales,
        color=colors,
        edgecolor='black',
        width=0.3)

plt.title("Monthly Sales 2024")
plt.xlabel("Month")
plt.ylabel("Sales (₹)")
plt.show()
```

Here we added different colors to diff bars using list. If any bar gets miss then it gets the first color.

```python
# EX 3
# Time spend in social media
name = ['ig', 'fb', 'x', 'sc', 'wp']
a = [45, 30, 20, 15, 60]
b = [60, 20, 45, 10, 30]
c = [25, 40, 10, 50, 25]
d = [35, 15, 30, 25, 40]

plt.bar(name, a, color='red')
plt.bar(name, b, color='green')
plt.bar(name, c, color='yellow')
plt.show()
```

```python
# better version (multiple bar chart or grouped bar chart)
import numpy as np
x = np.arange(len(name))

plt.bar(x, a, color='red', width=0.2)
plt.bar(x+0.2, b, color='green', width=0.2)
plt.bar(x+0.4, c, color='yellow', width=0.2)
plt.xticks(x, name)
plt.show()
```

1. Simple bar chart (vertical)
2. horizontal bar chart
3. multiple bar chart
4. stacked bar chart

```python
# Stacked ex
name = ['a', 'b', 'c']
a = [30, 50, 70]
b = [15, 25, 50]

plt.bar(name, a)
plt.bar(name, b)
plt.show()
```

```python
# Labels on bars

categories = ['Mango', 'Apple', 'Banana', 'Orange']
values     = [30, 20, 25, 15]

bars = plt.bar(categories, values, color='steelblue', edgecolor='black')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x position (center of bar)
        height + 0.5,                        # y position (just above bar)
        str(height),                         # the text to display
        ha='center',                         # horizontal alignment
        va='bottom',                         # vertical alignment
        fontsize=12
    )

plt.show()
```

```python
# Practice
months    = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_A = [500, 600, 550, 700, 650]
product_B = [300, 350, 400, 300, 500]
product_C = [200, 250, 150, 300, 350]
```

```python
# simple
plt.bar(months, product_A)
plt.show()
```

```python
# Stacked
plt.bar(months, product_A)
plt.bar(months, product_B)
plt.bar(months, product_C)
plt.show()
```

```python
# Color based on value
colors = ['green' if sales > 600 else 'red' for sales in product_A]


plt.bar(months, product_A, color=colors)
plt.show()
```

```python
# Grouped
x = np.arange(len(months))
plt.bar(x+0.2, product_A, width=0.2, color='orange', alpha=0.5)
plt.bar(x+0.4, product_B, width=0.2, color='red', alpha=0.5)
plt.bar(x+0.6, product_C, width=0.2, color='green', alpha=0.5)
plt.show()
```

```python
# Grouped
x = np.arange(len(months))
a = plt.bar(x+0.2, product_A, width=0.2, color='orange', alpha=0.6)
b = plt.bar(x+0.4, product_B, width=0.2, color='red', alpha=0.6)
c = plt.bar(x+0.6, product_C, width=0.2, color='pink', alpha=0.7)

for bar in a:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,   # x position (center of bar)
        height + 0.5,                        # y position (just above bar)
        str(height),                         # the text to display
        ha='center',                         # horizontal alignment
        va='bottom',                         # vertical alignment
        fontsize=8
    )
for bar in b:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        str(height),
        ha='center',
        va='bottom',
        fontsize=8
    )
for bar in c:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        str(height),
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.show()
```
