# Line plot

Line plot is the most basic and commonly used chart.  
It connects the data points with a line.

```python
import matplotlib.pyplot as plt
```

**Basic plot**

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 40]
```

```python
plt.plot(x, y)
plt.show()
```

Its just a plane chart. But we can custasmize almost everything in it.

---

|Parameter|Info|Ex|
|-|-|-|
|color|line color|'red', 'green', '#FFFFFF'|
|linewidth|thickness of connecting line|1, 2, 2.5|
|linestyle|line style|'-', '--', ':', '-.'|
|marker|data point shape|'o', 's', '*'|
|markersize|data point size|5, 8, 9.7|

---

`plt.title("title_name")` - Gives name to the chart  
`plt.xlabel("xlabel_name")` - Name for x axis   
`plt.ylabel("ylabel name")` - name for y axis  

```python
# EX
plt.plot(x, 
         y,
        color='red',
        linewidth=2,
        linestyle='--',
        marker='*',
        markersize=10)
plt.title("Name")
plt.xlabel("x_name")
plt.ylabel("y_name")
plt.show()
```

```python
# EX2
days = [1, 2, 3, 4, 5, 6, 7]
actual  = [2, 3, 1, 4, 3, 5, 2]
target  = [4, 4, 4, 4, 4, 4, 4]

plt.plot(days, actual, color='orange', marker='o', label='Actual')
plt.plot(days, target, color='red', linestyle='--', label='Target')

plt.title("Study Hours")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.legend()   # shows the labels for each plot
plt.show()
```

***IMP***

`plt.legend()` — whenever you have more than one line, always add a legend so the reader knows which line is which.
